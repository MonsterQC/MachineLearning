```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
plt.style.available
plt.style.use("seaborn-darkgrid")
```

### Mô hình Linear Regression sử sụng thư viện Sklearn

```python
X_train = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T # feature: Chiều cao
y_train = np.array([[49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T # Target: cân nặng

plt.plot(X_train, y_train, "ro", label="Visualize data")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.axis("equal")


lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)

X_model = np.array([[145, 190]]).T
y_model = lr.coef_*X_model + lr.intercept_

plt.plot(X_model, y_model)
plt.legend();
```

```python
X_test = np.array([[156, 162]])
y_predicts = lr.coef_*X_test + lr.intercept_

print(y_predicts)
```
