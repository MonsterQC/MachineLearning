```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use("seaborn-whitegrid")
```

### Sử dụng mô hình Linear Regression dự đoán target phụ thuộc vào features 

```python
X_train = [147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183] # Chiều cao
y_train = [49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68] # Cân nặng

plt.plot(X_train, y_train, "ro", label="Visualize data")

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Plot Demo")
# plt.axis("tight")
plt.axis("equal")


X_train = np.array([X_train]).transpose()
y_train = np.array([y_train]).T
ones = np.ones((X_train.shape[0], 1), dtype=np.int8)
X = np.concatenate((X_train, ones), axis=1)

# Công thức Linear Regression
x = np.linalg.inv(X.transpose().dot(X)).dot(X.transpose().dot(y_train))

X_model = np.array([[145,185]]).T
y_model = x[0][0]*X_model + x[1][0]
plt.plot(X_model, y_model, label="Fitting Line")

plt.legend();
```

```python
# dự đoán cân nặng đối với chiều cao 156 cm, 162 cm

X_predicts = np.array([[156, 162]])
y_predicts = x[0][0]*X_predicts + x[1][0]

print(y_predicts)
```
