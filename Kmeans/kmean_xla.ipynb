```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

### Ứng dụng thuật toán Kmeans trong xử lý ảnh -> Hình minh họa

```python
# read image
original_img = plt.imread("a.jpg")
height = original_img.shape[0]
width = original_img.shape[1]

img = original_img.reshape(height*width, 3)
print(img)
```

```python
kmeans = KMeans(n_clusters=2).fit(img) # Hình mới dc chọn từ 2 màu đặc trưng nhất
labels = kmeans.predict(img)
clusters = kmeans.cluster_centers_

print(clusters)
```

```python
img2 = np.zeros_like(img)
for i in range(len(img2)):
    img2[i] = clusters[labels[i]]

img2 = img2.reshape(height, width, 3)

fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(14,6))

ax1.imshow(original_img);
ax2.imshow(img2);
```
